```python
wt= pd.read_csv(r'C:\Users\16479\Downloads\weatherstats_toronto_daily.csv')
```

```python
wt['date'] = pd.to_datetime(wt['date'])  # Convert date column to datetime if needed

# Filter rows based on the date range
start_date = pd.to_datetime('2004-01-01')
end_date = pd.to_datetime('2022-12-31')
filtered_wt = wt[(wt['date'] >= start_date) & (wt['date'] <= end_date)]
```

```python
wt.info()
```

```python
desired_columns = [col for col in wt.columns if col.startswith('avg') and 'hourly' not in col]
desired_columns += ['date','snow', 'precipitation', 'rain']
filtered_wt = wt[desired_columns]
```

```python
fields_to_remove = ['avg_health_index','avg_dew_point', 'avg_pressure_station', 'avg_pressure_sea', 'avg_cloud_cover_4', 'avg_cloud_cover_8', 'avg_cloud_cover_10', 'avg_dew_point']
filtered_wt = filtered_wt.drop(fields_to_remove, axis=1)
```

```python
# Fill null values with forward fill strategy
filtered_wt['avg_temperature'].fillna(method='ffill', inplace=True)
filtered_wt['avg_relative_humidity'].fillna(method='ffill', inplace=True)
filtered_wt['avg_wind_speed'].fillna(method='ffill', inplace=True)
filtered_wt['avg_visibility'].fillna(method='ffill', inplace=True)
filtered_wt['snow'].fillna(method='ffill', inplace=True)
filtered_wt['precipitation'].fillna(method='ffill', inplace=True)
filtered_wt['rain'].fillna(method='ffill', inplace=True)
```

```python
null_counts = filtered_wt.isnull().sum()
print(null_counts)
```

```python
filtered_wt.info()
```

```python
df['OCC_DATE'] = pd.to_datetime(df['OCC_DATE']).dt.date
filtered_wt['date'] = pd.to_datetime(filtered_wt['date']).dt.date

# Merge the datasets based on the common date column
merged_df = pd.merge(df, filtered_wt, left_on='OCC_DATE', right_on='date')

merged_df.info()
```

```python
merged_df.insert(merged_df.columns.get_loc('INJURIES') + 1, 'TOTAL_CASUALTIES', merged_df['DEATH'] + merged_df['INJURIES'])
```

```python
correlation = merged_df[['avg_temperature', 'avg_wind_speed', 'avg_visibility', 'snow', 'precipitation', 'rain', 'TOTAL_CASUALTIES']].corr()
correlation 
```

```python
import statsmodels.api as sm

# Define the predictor variables (weather fields)
X = merged_df[['avg_temperature', 'avg_wind_speed', 'avg_visibility', 'snow', 'precipitation', 'rain']]

# Add a constant column to the predictor variables
X = sm.add_constant(X)

# Define the target variable (TOTAL_CASUALTIES)
y = merged_df['TOTAL_CASUALTIES']

# Fit the OLS model
model = sm.OLS(y, X)
results = model.fit()

# Print the summary of the regression results
print(results.summary())
```

```python
merged_df['TOTAL_CASUALTIES'] = pd.to_numeric(merged_df['TOTAL_CASUALTIES'])
```

```python
import statsmodels.api as sm

# Create new numerical fields based on 'OCC_TIME_RANGE' and 'OCC_DOW'
time_range_mapping = {
    'Night': 1,
    'Evening': 2,
    'Afternoon': 3,
    'Morning': 4
}

dow_mapping = {
    'Monday': 1,
    'Tuesday': 2,
    'Wednesday': 3,
    'Thursday': 4,
    'Friday': 5,
    'Saturday': 6,
    'Sunday': 7
}

merged_df['time_range_num'] = merged_df['OCC_TIME_RANGE'].map(time_range_mapping)
merged_df['dow_num'] = merged_df['OCC_DOW'].map(dow_mapping)

# Perform OLS regression
X = merged_df[['time_range_num', 'dow_num']]
X = sm.add_constant(X)  # Add a constant term to the model
y = merged_df['TOTAL_CASUALTIES']


model = sm.OLS(y, X)
results = model.fit()

# Print the summary of the regression results
print(results.summary())
```
